#comparing their scores

The R-squared (R2) score, also known as the coefficient of determination, is a statistical measure used to assess the goodness of fit of a regression model. It indicates the proportion of the variance in the dependent variable (target) that is explained by the independent variables (features) in the model. In other words, R2 quantifies how well the model's predictions match the actual observed values.

The R2 score is a value between 0 and 1. Here's what different R2 values imply:

* R2 = 0: The model explains none of the variability in the target variable. It's as good as predicting the mean of the target for every observation.
* 0 < R2 < 1: The model explains a portion of the variability in the target variable. The closer R2 is to 1, the better the model's predictions align with the actual data.
* R2 = 1: The model perfectly explains all the variability in the target variable. It predicts every observation exactly.
Mathematically, the R2 score is calculated as:

R2 = 1 - (SSR / SST)

Where:

* SSR (Sum of Squared Residuals) is the sum of the squared differences between the predicted values and the actual values.
* SST (Total Sum of Squares) is the sum of the squared differences between the actual values and the mean of the actual values.

The R2 score ranges from 0 to 1, but it's important to note that a high R2 score doesn't necessarily mean the model is good. A high R2 score might indicate overfitting if the model is capturing noise in the data instead of true underlying relationships. Therefore, it's important to consider other evaluation metrics, cross-validation, and domain knowledge when assessing the quality of a regression model.

#generating polynomial features

generating polynomial features means it takes the existing 10 dimensions , understands the relation between them and then creates more dimensions based on the relations between them.

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True)
```

degree=2 here means that a dimension can be created which is the original dimension raised to power 2 but not more than that.

interaction_only=True means that only the dimensions showing some kind of interaction or correlation between them will be taken.



```python
x_poly=poly.fit_transform(x_scaled)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.30,random_state=1)
x_train.shape
```
